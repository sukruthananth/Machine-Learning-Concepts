```python
# we can use gridsearchcv() to find the best parameters using cross validation -  grid_search.best_estimator_


from sklearn.model_selection import GridSearchCV

from sklearn.model_selection import train_test_split

from sklearn.datasets import load_iris

iris = load_iris()

from sklearn.svm import SVC

params=[{'Kernel':['rbf'],'C': [0.001, 0.01, 0.1, 1, 10, 100],'gamma': [0.001, 0.01, 0.1, 1, 10, 100]},{'kernel': ['linear'],'C': [0.001, 0.01, 0.1, 1, 10, 100]}]

gridsearch = GridSearchCV(SVC(), param_grid=params, cv = 5)

X_train, X_test, y_train, y_test = train_test_split(
        iris.data, iris.target, random_state=0)

gridsearch.fit(X_train, y_train)

gridsearch.cv_results_
```

```python
# we can use dummy classifier to find out if the output is not balanced
from sklearn.datasets import load_digits 
digits = load_digits()
y = digits.target == 9

from sklearn.dummy import DummyClassifier

X_train, X_test, y_train, y_test = train_test_split(
digits.data, y, random_state=0)

dummy = DummyClassifier().fit(X_train, y_train)

pred_most_frequent = dummy.predict(X_test)
pred_most_frequent
```
